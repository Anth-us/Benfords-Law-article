```python
import matplotlib.pyplot as plt
import numpy as np

# Generate Benford distribution data
numbers = np.arange(1, 10)
benford_probs = np.log10(1 + 1/numbers)

plt.figure(figsize=(6, 6))
plt.bar(numbers, benford_probs, color='black')
plt.axis('off')
plt.savefig('bendford-distribution.png')
plt.show()
```
